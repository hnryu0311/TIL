#### 연습문제 9.3.1

```python
xx = np.linspace(0, 1, 1000)

# 상품 A의 사후 확률
a0, b0 = 2+1, 1+1
plt.plot(xx, sp.stats.beta(a0, b0).pdf(xx), c='r', ls = '--', label = '상품 A의 사후분포')

#상품 B의 사후 확률
a1, b1 = 60+1, 40+1
plt.plot(xx, sp.stats.beta(a1, b1).pdf(xx), c='g', ls = '-', label = '상품 B의 사후분포')

plt.title('베르누이분포의 모수를 베이즈 추정법으로 추정한 결과')
```

#### 연습문제 9.3.2

```python
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data[:, 0]
N = len(x)
mu = np.mean(x)
sigma2 = np.var(x)
print("sample mean:", mu)
print("sample variance:", sigma2)
sns.distplot(x, kde=False, fit=sp.stats.norm)
plt.vlines(x=mu, ymin=0, ymax=0.6, ls="--")
plt.vlines(x=mu+2*np.sqrt(sigma2), ymin=0, ymax=0.6, ls=":")
plt.vlines(x=mu-2*np.sqrt(sigma2), ymin=0, ymax=0.6, ls=":")
plt.show()
```
